**Why this dataset:** IHDP is semi-synthetic — real covariates from the Infant Health and Development Program, but a simulated outcome with a *known* true individual treatment effect. **Technique focus:** double/debiased ML (DML) for the ATE, causal forests or EconML's `CausalForestDML`/`LinearDML` for heterogeneous effects (CATE).
